**Note!** : 

1. Canvas 0,0 starts on the top left corner. Bounding box: [100, 50, 250, 300]. Here, the top-left corner is at (100, 50) and the bottom-right corner is at (250, 300). This would represent a rectangle with a width of 250 - 100 = 150 pixels and a height of 300 - 50 = 250 pixels.

#### Simple Example


```python
import yaml
import os
from pathlib import Path
from IPython.display import Image

new_path = "/home/jupyter/InstantPixel/GLIGEN/"
os.chdir(new_path)

def modify_yaml(custom_prompt: str, phrase, box, output_folder):
    # Load the data
    with open("prompt.yaml", 'r') as file:
        data = yaml.safe_load(file)

    # Modify a value
    data['prompt'] = custom_prompt  

    # Add a new key-value pair
    data['location'] = box  
    
    # Add a new key-value pair
    data['phrase'] = phrase  
    
    data['output_folder'] = output_folder

    # Save the changes
    with open("prompt.yaml", 'w') as file:
        yaml.dump(data, file)

    print("YAML file modified successfully!")


```

```python
# new_path = Path("simple_example_output")
# new_path.mkdir(parents=True, exist_ok=True)

prompt = "Northern lights lighting up the night sky with a reindeer" 
phrase = ['northern lights', 'a reindeer']
box = [ [0.0,0.0,1.0,0.3], [0.7,0.8,0.8,0.9] ]
output_folder = "simple_example_output"
modify_yaml(prompt, phrase, box, output_folder)

print("Executing gligen code...")
! python gligen_inference.py
```
